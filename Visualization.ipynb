```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head().iloc[:, :7]
```

```python
plt.figure(num=None, figsize=(12, 8), dpi=80, facecolor='w', edgecolor='k')

plt.plot(df['Unnamed: 1'], df['RPQ SIMPLE']/1000, '-o', label = 'Non Optimized')
plt.plot(df['Unnamed: 1'], df['OPT']/1000, '-o', label = 'Optimized')


plt.yticks(np.arange(0, max(df['RPQ SIMPLE']/1000)+0.25, 0.25))
plt.xticks(np.arange(0, max(df['Unnamed: 1'])+500, 500));


plt.xlabel('Number of edges', size = 15)
plt.ylabel('Execution time (sec)', size = 15)
plt.legend(loc = 'best')
plt.grid()
plt.title('RPQ: a+/b', size = 20);
```
